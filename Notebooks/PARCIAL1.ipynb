```python
import pandas as pd 
import numpy as np
from pathlib import Path 
```

```python
#Caracteristicas generales, seguridad social y eduacciÃ³n 
#Listas a los conjuntos de datos
url_cgsse = "https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/cgsse.CSV"
url_ocupados = "https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Ocupados.CSV"
url_ft = "https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Fuerza%20de%20trabajo.CSV"
url_desocupados = "https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/No%20ocupados.CSV"

# CGSSE
df_cgsse = pd.read_csv(url_cgsse, encoding='latin-1', sep=';', 
                       low_memory=False)

#Ocupados
df_ocupados=pd.read_csv(url_ocupados,encoding='latin-1',sep=';',
                        low_memory=False)

#Fuerza de trabajo
df_ft=pd.read_csv(url_ft,
                  encoding='latin-1',sep=';',low_memory=False)
#No ocupados
df_desocupados=pd.read_csv(url_desocupados,encoding='latin-1',
                           sep=';',low_memory=False)
```

1. Para el total nacional calcule :
 
-poblaciÃ³n total
-poblaciÃ³n en edad de trabajar
-poblaciÃ³n que no estÃ¡ en edad de trabajar
-poblaciÃ³n econÃ³micamente activa
-poblaciÃ³n econÃ³micamente inactiva
-tasa de ocupaciÃ³n
-tasa de desempleo


```python
#POBLACIÃN TOTAL 
PT =  df_cgsse['FEX_C18'].sum()
print(f' La poblaciÃ³n total de referencia para Colombia es de {PT}')
```

```python
#POBLACION EN EDAD DE TRABAJAR 
PET = df_cgsse[(df_cgsse['P6040']>=15) & (df_cgsse['P6040']<=65)]['FEX_C18'].sum()
print(f" La poblaciÃ³n en edad de trabajar para Colombia es de {PET} ")
```

```python
#POBLACIÃN QUE NO ESTA EN EDAD DE TRABAJAR 
PENT = df_cgsse[(df_cgsse['P6040'] < 15) | (df_cgsse['P6040'] > 65)]['FEX_C18'].sum()  
print(f" La poblaciÃ³n que no esta en edad de trabajar para Colombia es de {PENT} ")
```

```python
#POBLACIÃN ECONOMICAMENTE ACTIVA
PEA=df_ft[(df_ft['FT']==1)]['FEX_C18'].sum()
print(f'PoblaciÃ³n Economicamente Activa {PEA}')
```

```python
#POBLACIÃN ECONOMICAMENTE INACTIVA
PEI=df_ft[(df_ft['FFT']==1)]['FEX_C18'].sum()
print(f'PoblaciÃ³n Economicamente Inactiva {PEI}')
```

```python
#TASA DE OCUPADOS
O=df_ocupados[(df_ocupados['OCI']==1)]['FEX_C18'].sum()
print(f'PoblaciÃ³n Ocupada es de {O}')
TO = O/PET
print(f'Tasa de Ocupacion: {TO:.2%}')
```

```python
#TASA DE DESEMPLEO
D=df_desocupados[(df_desocupados['DSI']==1)]['FEX_C18'].sum()
print(f'PoblaciÃ³n Desocupada {D}')
TD = D/PEA
print(f'Tasa de Desempleo: {TD:.2%}')
```

2. Calcule las tasas de desempleo para hombres y mujeres en Risaralda, Caldas , Tolima y QuindÃ­o. Â¿CuÃ¡l es el departamento con mayor y menor tasa de desempleo de los 4 ? 

```python
df_d = pd.read_csv('https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/No%20ocupados.CSV'
                       ,encoding='latin',sep=';')
```

```python
#Primero hay que anexar la p3271 a la tabla de desempleados
df_d = pd.merge(df_d, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'P3271','P6040']], 
                        on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')

df_ft = pd.merge(df_ft, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'P3271','P6040']], 
                        on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')
```

```python
# 1. Calcular cantidad de hombres desempleados en Risaralda
HDQ = df_d[(df_d['DSI']==1) & (df_d['DPTO']==66) & (df_d['P3271']==1)]['FEX_C18'].sum()
print(f'Hombres Desempleados en Risaralda: {HDQ}')

# 2. Calcular cantidad de hombres en la PEA del Risaralda
PEAHQ = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==66) & (df_ft['P3271']==1) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de hombres en el Risaralda: {PEAHQ}')

# 3. Calcular la tasa de desempleo en hombres del Risaralda
TDHQ = HDQ/PEAHQ
print(f'Tasa de Desempleo en hombres del Risaralda: {TDHQ:.2%}')
```

```python
# 1. Calcular cantidad de mujeres desempleadas en Risaralda
HDQ = df_d[(df_d['DSI']==1) & (df_d['DPTO']==66) & (df_d['P3271']==2)]['FEX_C18'].sum()
print(f'Mujeres Desempleadas en Risaralda: {HDQ}')

# 2. Calcular cantidad de mujeres en la PEA del Risaralda
PEAHQ = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==66) & (df_ft['P3271']==2) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de mujeres en el Risaralda: {PEAHQ}')

# 3. Calcular la tasa de desempleo en mujeres del Risaralda
TDHQ = HDQ/PEAHQ
print(f'Tasa de Desempleo en mujeres del Risaralda: {TDHQ:.2%}')
```
